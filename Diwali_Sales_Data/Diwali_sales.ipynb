```python
ax = sns.countplot(x = 'Gender', data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x = 'Gender', y = 'Amount', data=sales_gen)
```

**From the above graph we can see that most of the buyers are female and even the purvchasing power of females are greater than the men.**

## **Age**

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')
```

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age= df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x = 'Age Group', y = 'Amount', data =sales_age)
```

**From above graph we can see that the most of the buyers are from the age group between 26-35 years female**

# **State**

```python
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by = 'Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State', y = 'Orders', palette = 'Set1')
```

```python
sales_amt = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_amt, x = 'State', y = 'Amount', palette = 'Set2') 
```

```python
ax = sns.countplot(data = df, x = 'Marital_Status', palette='Set3')
sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
mrtl_stat = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = mrtl_stat, x = 'Marital_Status', y='Amount', hue = 'Gender')
```

**From above graph we can see that most of the buyers are married (women) and they have high purchasing power**

# **Occupation**

```python
sns.set(rc={'figure.figsize':(20,5)})
ax= sns.countplot(data = df, x = 'Occupation', hue='Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
occ_amt = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = occ_amt , x = 'Occupation', y='Amount', hue='Occupation')
```

**From above graph we can see that the most of the buyers are working in IT , Healthcare and Aviation sector**

# **Product Category**

```python
sns.set(rc={'figure.figsize':(20,5)})
ax= sns.countplot(data = df, x = 'Product_Category', hue='Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
prdct_cat = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = prdct_cat , x = 'Product_Category', y='Amount', hue='Product_Category')
```

**From above graph we can see that the most liked product is the food**

```python
prdct_id = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by = 'Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = prdct_id , x = 'Product_ID', y='Orders', hue='Product_ID')
```

**Married women of age group 26-35 yrs from UP, Maharashtra, and Karnataka, working in IT, Healthcare and aviation are more likely to buy products from Food, Clothing and Electronics category.**
